```python
#Imports das bibliotecas
import matplotlib.pyplot as plt
from numpy import arange
import numpy as np

#Definição da função analítica do gráfico
def newt(y,u_0,mi,D):
    P = -8*mi*u_0/(D**2)
    
    A = -P*D/2
    
    u = P*(y**2)/(2*mi) + A*y/(mi)

    return u

#Velocidade inicial
u_0 = 2/3.6 # [m/s]          

#Diâmetro inicial
D = 0.02  # [m]

#Viscosidade inicial
mi = 4.4*10e-3   # []      ## si ##?

#Diâmetro inicial
t_i = 0 #[m]

#Diâmetro final
t_f = D #[m]

#Número de divisões
div = 1000

#Tamanho do passo
pas = (t_f-t_i)/div

#Parte de plotagem#
y = arange(t_i,t_f,pas) 

plt.figure(figsize=(7,7))

while u_0 <= 10:
    u = []

    for i in y:
        u.append(newt(i,u_0,mi,D))
        
    plt.plot(u, y,label = 'D = %.2f m e v = %.1f m/s'%(D,u_0))
    u_0 += 3.2

#plt.ylim(3,16)
#plt.xlim(7.5,13.5)
plt.xlabel('velocidade U'),plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()

```

```python
#Definição da função analítica do gráfico
def non_newt(y,n,u_0,mi,D):
    
    
    #P = -((2*u_0*(n+1)/(n*D))**(n))*(2*mi/D)
    P  = -(pow(2*u_0*(n+1)/(n*D),n)*(2*mi/D))
    
    A = -P*D/2
    
    #C1 = -n*A*((-A/mi)**(1/n))
    C1 = -n*A*pow((A/mi),(1/n))
    
           
    #u_ = n*(A + P*y)*(((A + P*y)/mi)**(1/n))
    
    if A + P*y >= 0: 
        u_ = n*(A + P*y)*pow((A + P*y)/mi,(1/n))
    else: 
        u_ = -n*(A + P*y)*pow(abs(A + P*y)/mi,(1/n))
   
    u = (u_ + C1)/(P*(n+1))

    return u

#Diâmtro final
D =   0.02 # [m]

#Viscosidade
mi =  4.4*10e-3 # [Pa]

# Viscosidade (Power law < 1 , Newtoniano = 1)
n =  0.5 # [?]

#Velocidade inicial
u_0 = 2/3.6 # [m/s]





#Parte de Plotagem#


plt.figure(figsize=(10,10))

plt.subplot(2, 2, 1).set_title("n = 0.5")

#Plot Power Law
while u_0 <= 10:
    
    t_i = 0
    t_f = D
    div = 1000
    pas = (t_f-t_i)/div

    y = arange(t_i,t_f,pas) 
    u = []

    for i in y:
        u.append(non_newt(i,n,u_0,mi,D))

    plt.plot(u, y,label = "v = %.1f m/s"%u_0)
    plt.xlabel('velocidade U')
    plt.ylabel('y')
    plt.legend()
    plt.grid()
        
        
    u_0 += 3.2




n =  1      
u_0 = 2/3.6 


#Plot Newtoniano
plt.subplot(2, 2, 2).set_title("n = 1")
while u_0 <= 10:
    
    t_i = 0
    t_f = D
    div = 1000
    pas = (t_f-t_i)/div

    y = arange(t_i,t_f,pas) 
    u = []

    for i in y:
        u.append(non_newt(i,n,u_0,mi,D))

    plt.plot(u, y,label = "v = %.1f m/s"%u_0)
    plt.xlabel('velocidade U')
    plt.ylabel('y')
    plt.legend()
    plt.grid()
        
        
    u_0 += 3.2
    

# Viscosidade (Power law < 1 , Newtoniano = 1)
n =  1.5 # []?

# Velocidade inicial
u_0 = 2/3.6 # [m/s]

#Plot Não-Newtoniano diferente do power law
plt.subplot(2, 2, 3).set_title("n = 1.5")
while u_0 <= 10:
    
    t_i = 0
    t_f = D
    div = 1000
    pas = (t_f-t_i)/div

    y = arange(t_i,t_f,pas) 
    u = []

    for i in y:
        u.append(non_newt(i,n,u_0,mi,D))

    plt.plot(u, y,label = "v = %.1f m/s"%u_0)
    plt.xlabel('velocidade U')
    plt.ylabel('y')
    plt.legend()
    plt.grid()
        
        
    u_0 += 3.2
    

plt.show()


```
